# **TURUNAN NUMERIK METODE BEDA HINGGA**

**Metode Beda Hingga Maju**

Turunan $f(x)$ didekati dengan $f'(x) \approx \dfrac{f(x + h) - f(x)}{h}$ untuk suatu nilai $h$ yang cukup kecil, dan turunan $f'(x)$ didekati dengan $f''(x) \approx \dfrac{f(x + 2h) - 2f(x + h) + f(x)}{h^2}$ untuk suatu nilai $h$ yang cukup kecil.

**Metode Beda Hingga Mundur**

Turunan $f(x)$ didekati dengan $f'(x) \approx \dfrac{f(x) - f(x - h)}{h}$ untuk suatu nilai $h$ yang cukup kecil, dan turunan $f'(x)$ didekati dengan $f''(x) \approx \dfrac{f(x - 2h) - 2f(x - h) + f(x)}{h^2}$ untuk suatu nilai $h$ yang cukup kecil.

**Metode Beda Hingga Tengah**

Turunan $f(x)$ didekati dengan $f'(x) \approx \dfrac{f(x + h) - f(x - h)}{2h}$ untuk suatu nilai $h$ yang cukup kecil, dan turunan $f'(x)$ didekati dengan $f''(x) \approx \dfrac{f(x + h) - 2f(x) + f(x - h)}{h^2}$ untuk suatu nilai $h$ yang cukup kecil.

## **ALGORITMA**

Contoh :
Diberikan fungsi $f$ dengan
$$
f(x) = x^3 + 4x + 4
$$
Dengan menggunakan Metode Beda hingga maju, mundur, dan tengah, tentukan $f'(0,5)$ dan $f''(0,5)$ dengan $h=0,25$.

```python
import numpy as np

#Definisikan Fungsi
def F(x):
    return x**3 + 4*x + 4

#Turunan Analitiknya
#Turunan Pertama
def f(x):
    return 3*x**2 + 4
#Turunan Kedua
def g(x):
    return 6*x
```

```python
# UNTUK TURUNAN PERTAMA

#Metode Beda Hingga Maju
def maju1(F, x, h):
    return (F(x + h) - F(x))/h

#Metode Beda Hingga Mundur
def mundur1(F, x, h) :
    return (F(x) - F(x - h))/h

#Metode Beda Hingga Tengah
def tengah1(F, x, h) :
    return (F(x + h) - F(x - h))/(2*h)

print("Dengan metode beda hingga maju diperoleh f'({}) mendekati {}".format(0.5, maju1(F, 0.5, 0.25)))
print("Dengan metode beda hingga mundur diperoleh f'({}) mendekati {}".format(0.5, mundur1(F, 0.5, 0.25)))
print("Dengan metode beda hingga tengah diperoleh f'({}) mendekati {}".format(0.5, tengah1(F, 0.5, 0.25)))
print("Galat metode beda hingga maju : ", abs(f(0.5) - maju1(F, 0.5, 0.25)))
print("Galat metode beda hingga mundur : ", abs(f(0.5) - mundur1(F, 0.5, 0.25)))
print("Galat metode beda hingga tengah : ", abs(f(0.5) - tengah1(F, 0.5, 0.25)))
```

```python
# UNTUK TURUNAN KEDUA

#Metode Beda Hingga Maju
def maju2(F, x, h):
    return (F(x + 2*h) - 2*F(x + h) + F(x))/(h**2)

#Metode Beda Hingga Mundur
def mundur2(F, x, h):
    return (F(x - 2*h) - 2*F(x - h) + F(x))/(h**2)

#Metode Beda Hingga Tengah
def tengah2(F, x, h):
    return (F(x + h) - 2*F(x) + F(x - h))/(h**2)

print("Dengan metode beda hingga maju diperoleh f''({}) mendekati {}".format(0.5, maju2(F, 0.5, 0.25)))
print("Dengan metode beda hingga mundur diperoleh f''({}) mendekati {}".format(0.5, mundur2(F, 0.5, 0.25)))
print("Dengan metode beda hingga tengah diperoleh f''({}) mendekati {}".format(0.5, tengah2(F, 0.5, 0.25)))
print("Galat metode beda hingga maju : ", abs(g(0.5) - maju2(F, 0.5, 0.25)))
print("Galat metode beda hingga mundur : ", abs(g(0.5) - mundur2(F, 0.5, 0.25)))
print("Galat metode beda hingga tengah : ", abs(g(0.5) - tengah2(F, 0.5, 0.25)))
```

```python
# PERBANDINGAN AKURASI TURUNAN NUMERIK SETIAP METODE

x = np.linspace(-2, 2, 100)
y_analitik = [f(titik) for titik in x]

y_maju = [maju1(F, titik, 0.5) for titik in x]
y_mundur = [mundur1(F, titik, 0.5) for titik in x]
y_tengah = [tengah1(F, titik, 0.5) for titik in x]

import matplotlib.pyplot as plt

plt.figure(figsize = (12, 7))
plt.plot(x, y_analitik, color = 'blue', label = "Solusi Analitik" )
plt.scatter(x, y_maju, color = 'black', label = 'Solusi Metode Beda Hingga Maju')
plt.scatter(x, y_mundur, color = 'green', label = 'Solusi Metode Beda Hingga Mundur')
plt.scatter(x, y_tengah, color = 'orange', label = 'Solusi Metode Beda Hingga Tengah')
plt.title("Grafik Perbandingan Solusi Analitik dengan Solusi Numerik")
plt.legend()
plt.grid()
plt.show()
```

# **METODE KOEFISIEN TAK TENTU**

Misalkan kita ingin mencari rumus untuk turunan pertama, $f'(x_0)$, menggunakan tiga titik: $x_0$, $x_0-h$, dan $x_0-2h$.
1. Diasumsikan rumusnya berbentuk:
   $$
   f'(x_0) \approx A \cdot f(x_0) + B \cdot f(x_0-h) + C \cdot f(x_0-2h)
   $$
   Di sini, $A$, $B$, dan $C$ adalah "koefisien tak tentu" yang akan dicari.

2. Ekspansikan setiap $f(x)$ di sekitar titik $x_0$ menggunakan Deret Taylor, diperoleh
   - $f(x_0) = f(x_0)$
   - $f(x_0-h) = f(x_0) - hf'(x_0) + \frac{h^2}{2}f''(x_0) - \frac{h^3}{6}f'''(x_0) + \dots$
   - $f(x_0-2h) = f(x_0) - 2hf'(x_0) + \frac{(2h)^2}{2}f''(x_0) - \frac{(2h)^3}{6}f'''(x_0) + \dots$

3. Substitusikan ekspansi Taylor ke dalam asumsi awal, diperoleh
   $$
   \begin{align*}f'(x_0)  & \approx A \cdot f(x_0) + B \cdot f(x_0-h) + C \cdot f(x_0-2h) \\
   & = A \cdot \left[f(x_0)\right] + B \cdot \left[f(x_0) - hf'(x_0) + \frac{h^2}{2}f''(x_0) - \dots\right] + C \cdot \left[f(x_0) - 2hf'(x_0) + 2h^2 f''(x_0) - \dots\right] \\
   & = (A+B+C) \cdot f(x_0) + h(-B - 2C) \cdot f'(x_0) + \frac{h^2}{2}(B + 4C) \cdot f''(x_0) + \dots
   \end{align*}
   $$

4. Karena yang dicari $f'(x_{0})$, maka dipilih koefisien $f'(x_{0})$ adalah $1$ dan yang lain $0$. Didapat sistem persamaan linear untuk $A$, $B$, dan $C$
   $$
   \begin{align*}
   A + B + C  & = 0 \\
   h(-B - 2C)  & = 1 \\
   \frac{h^2}{2}(B + 4C)  & = 0
   \end{align*}
   $$
   Karena banyak persamaan sudah cukup, suku yang lain dipotong.
   Sistem persamaan tersebut dapat ditulis dalam bentuk matriks
   $$
   \begin{bmatrix}
   1 & 1 & 1 \\
   0 & -h & -2h \\
   0 & \frac{h^2}{2} & 2h^2
   \end{bmatrix}\begin{bmatrix}
   A \\
   B \\
   C
   \end{bmatrix}=\begin{bmatrix}
   0 \\
   1 \\
   0
   \end{bmatrix}
   $$
   $$
   Mc = b
   $$

5. Dengan menyelesaikan sistem tersebut, diperoleh koefisiennya
   - $A = \dfrac{3}{2h}$
   - $B = -\dfrac{2}{h}$
   - $C = \dfrac{1}{2h}$
   
   Sehingga diperoleh rumus pendekatan turunannya
   $$
   \begin{align*}
   f'(x_0)  & \approx \frac{3}{2h}\cdot f(x_0) - \frac{2}{h}\cdot f(x_0-h) + \frac{1}{2h}\cdot f(x_0-2h) \\
   & =\frac{3f(x_0) - 4f(x_0-h) + f(x_0-2h)}{2h}
   \end{align*}
   $$

Misalkan kita ingin mencari rumus untuk turunan ke-$i$, $f^{(i)}(x_0)$, menggunakan nilai fungsi $f(x_1)$, $f(x_2)$, $\dots$ $f(x_n)$, dengan $n>i$ dan
* $x_1 = x_0 + h_1$
* $x_2 = x_0 + h_2$
    
    $\vdots$
* $x_n = x_0 + h_n$

Diasumsikan rumusnya berbentuk
$$
f^{(i)}(x_0) \approx A_1 \cdot f(x_1) + A_2 \cdot f(x_2) + \dots + A_n \cdot f(x_n)
$$

Sehingga sistem persamaannya akan berbentuk
$$
\begin{bmatrix}
1 & 1 & \cdots & 1 \\
h_1 & h_2 & \cdots & h_n \\
\dfrac{h_1^2}{2} & \dfrac{h_2^2}{2} & \cdots & \dfrac{h_n^2}{2} \\
\vdots & \vdots &  & \vdots \\
\dfrac{h_1^{i}}{i!} & \dfrac{h_2^{i}}{i!} & \cdots & \dfrac{h_n^{i}}{i!} \\
\vdots & \vdots &  & \vdots \\
\dfrac{h_1^{n-1}}{(n-1)!} & \dfrac{h_2^{n-1}}{(n-1)!} & \cdots & \dfrac{h_n^{n-1}}{(n-1)!}
\end{bmatrix}
\begin{bmatrix}
A_1 \\
A_2 \\
\vdots \\
A_n
\end{bmatrix}
=
\begin{bmatrix}
0 \\
0 \\
0 \\
\vdots \\
1 \\
\vdots \\
0
\end{bmatrix}
$$

# Contoh Soal

Diberikan data pada tabel berikut.
|   $x$   |    $y$     |
|:-------:|:----------:|
| $0.10$  | $4.2100$   |
| $0.15$  | $5.8225$   |
| $0.20$  | $7.4400$   |

Dengan menggunakan Metode Koefisien Tak Tentu, tentukan nilai $f'(0,1)$.


```python
import numpy as np
import math

def find_coefficients(points, derivative_order):
    # Jumlah titik (dan juga jumlah koefisien)
    n = len(points)
    if n <= derivative_order:
        raise Exception("Jumlah titik harus lebih dari orde turunan")

   # Buat matriks M (sistem persamaan)
    M = np.zeros((n, n))
    for i in range(n):
        for j in range(n):
            M[i, j] = points[j] ** i / math.factorial(i)

    # Buat vektor b (rhs)
    b = np.zeros(n)
    b[derivative_order] = 1

    # Selesaikan sistem linear M * c = b untuk mencari koefisien c
    coefficients = np.linalg.solve(M, b)

    return coefficients

print(find_coefficients([0, -1, -2], 1))
```

```python
def undetermined_coeff(x, y, derivative_order, target):
    sum = 0
    coeffs = find_coefficients(np.array(x) - target, derivative_order)
    for i in range(len(x)):
        sum += coeffs[i]*y[i]
    return sum
```

```python
x = [0.1, 0.15, 0.2]
y = [4.21, 5.8225, 7.44]
print(undetermined_coeff(x, y, 1, 0.1))
```

# **ANALISIS SENSITIVITAS**

## 1. Galat Pemotongan (Truncation Error)

Metode beda hingga tengah:
$$
f'(x_0) \approx \frac{f(x_0+h) - f(x_0-h)}{2h}
$$
Dilihat deret taylor nya    
- $f(x_0+h) = f(x_0) + hf'(x_0) + \frac{h^2}{2}f''(x_0) + \frac{h^3}{6}f'''(x_0) + \frac{h^4}{24}f^{(4)}(x_0) + \frac{h^5}{120}f^{(5)}(x_0) + \dots$
- $f(x_0-h) = f(x_0) - hf'(x_0) + \frac{h^2}{2}f''(x_0) - \frac{h^3}{6}f'''(x_0) + \frac{h^4}{24}f^{(4)}(x_0) - \frac{h^5}{120}f^{(5)}(x_0) + \dots$

Substitusi ke rumus, diperoleh
$$
\begin{align*}
  \frac{f(x_0+h) - f(x_0-h)}{2h} & = \frac{\left[f(x_0) + hf'(x_0) + \frac{h^2}{2}f''(x_0) + \frac{h^3}{6}f'''(x_0) + \dots \right]-\left[f(x_0) - hf'(x_0) + \frac{h^2}{2}f''(x_0) - \frac{h^3}{6}f'''(x_0) + \dots \right]}{2h} \\
  & = f'(x_0) + \underbrace{ \frac{h^2}{6}f'''(x_0) + O(h^4)  }_{ \text{galat pemotongan} }
\end{align*}
$$
Sehingga
$$
\left|E_{trunc}\right|=\left|\frac{h^2}{6}f'''(x_0) + O(h^4)\right|= \frac{h^2}{6}|f'''(\xi)| = Ch^2
$$
Untuk suatu $\xi$ dan $C$

---

## 2. Galat Pembulatan (Round-off Error)

Diperhatikan rumus yang sama: $f'(x_0) \approx \frac{f(x_0+h) - f(x_0-h)}{2h}$.

Ketika komputer menghitung $f(x_0+h)$, nilai yang disimpan bukanlah $f(x_0+h)$, tapi nilai hasil pembulatan yang paling dekat, sebut saja $\hat{f}(x_0+h)$.
- $\hat{f}(x_0+h) = f(x_0+h) + e_1$
- $\hat{f}(x_0-h) = f(x_0-h) + e_2$

dengan $e_1$ dan $e_2$ adalah galat pembulatan.

Diperoleh
$$
\begin{align*}
\frac{f(x_0+h) - f(x_0-h)}{2h}  & = \frac{(\hat{f}(x_0+h)-e_{1}) - (\hat{f}(x_0+h)-e_{2})}{2h} \\
 & = \frac{\hat{f}(x_0+h)-\hat{f}(x_0+h)}{2h} - \frac{e_{1}-e_{2}}{2h} \\
\end{align*}
$$
$$
\implies \frac{\hat{f}(x_0+h) - \hat{f}(x_0-h)}{2h} = \frac{f(x_0+h) - f(x_0-h)}{2h} + \underbrace{ \frac{e_{1}-e_{2}}{2h} }_{ \text{galat pembulatan} }
$$
Sehingga
$$
|E_{roff}|=\left|\frac{e_{1}-e_{2}}{2h}\right|\leq \frac{2\max\{|e_{1}|,|e_{2}|\}}{2h}\leq  \frac{\epsilon_{mach}}{h}
$$
Dengan $\epsilon_{mach}$ merupakan presisi mesin $\approx 10^{-16}$

### Galat total
$E = E_{trunc}+E_{roff}$

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    """Fungsi yang akan kita turunkan."""
    return np.sin(x)

def f_prime_exact(x):
    """Turunan analitik (eksak) dari f(x) untuk perbandingan."""
    return np.cos(x)

def central_difference(f, x, h):
    """Rumus beda hingga pusat untuk turunan pertama."""
    return (f(x + h) - f(x-h)) / (2*h)

# Titik di mana kita mengevaluasi turunan
x_val = np.pi / 4

# Hitung nilai turunan yang sebenarnya (eksak)
true_derivative = f_prime_exact(x_val)

# Buat rentang nilai h pada skala logaritmik
# Mulai dari 1e-17 (sangat kecil, dekat presisi mesin) hingga 1e0 (h=1)
h_values = np.logspace(-15, 0, 100)

# Hitung nilai turunan numerik untuk setiap h
numerical_derivatives = central_difference(f, x_val, h_values)

# Hitung galat absolut untuk setiap h
errors = np.abs(numerical_derivatives - true_derivative)

plt.figure(figsize=(10, 6))
plt.plot(h_values, errors, 'bo-', markersize=4, label='Galat Total Numerik')

# Menambahkan garis teoritis untuk ilustrasi
emach = np.finfo(float).eps # Presisi mesin, sekitar 2.2e-16

# Galat Pembulatan (Round-off) ~ emach / h
plt.plot(h_values, (emach / h_values), 'g--',
         label=r'Galat Pembulatan (O($\epsilon/h^2$))')

# Galat Pemotongan (Truncation) ~ C * h^2
# (Konstanta C dipilih agar pas secara visual. Disini dipilih C = 1)
plt.plot(h_values, (h_values**2), 'r--',
         label=r'Galat Pemotongan (O($h^2$))')

# Konfigurasi plot log-log
plt.xscale('log')
plt.yscale('log')

plt.title('Analisis Galat Turunan Numerik vs. Ukuran Langkah (h)', fontsize=14)
plt.xlabel('Ukuran Langkah (h)', fontsize=12)
plt.ylabel('Galat Absolut', fontsize=12)
plt.legend()
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
# plt.gca().invert_xaxis() # Membalik sumbu x agar h besar di kiri

print(f"Nilai eksak f'({x_val:.2f}) = {true_derivative}")
print(f"Epsilon (presisi mesin) = {emach}")
plt.show()
```

3. Buktikan bahwa, titik $x = -0,1678833571446$ merupakan maksimum lokal dari fungsi
$$
f(x) = e^x \frac{(\arccos(x))^2}{\cosh(x)}
$$
Hint: Ingat kembali materi Kalkukus I tentang titik ekstrem.


```python
import numpy as np
def F(x):
  return np.exp(x)*(np.arccos(x)**2)/(np.cosh(x))
def tengah1(F, x, h) :
    return (F(x + h) - F(x - h))/(2*h)

def tengah2(F, x, h):
    return (F(x + h) - 2*F(x) + F(x - h))/(h**2)

print("Dengan metode beda hingga tengah diperoleh f'({}) mendekati {}".format(-0.1678833571446, tengah1(F, -0.1678833571446, 0.001)))
print("Dengan metode beda hingga tengah diperoleh f''({}) mendekati {}".format(-0.1678833571446, tengah2(F, -0.1678833571446, 0.001)))
```
